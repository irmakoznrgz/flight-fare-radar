```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

flight_data_path = "../data/raw/flights.csv"
df = pd.read_csv(flight_data_path)

df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe(include='all').T
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], kde=True, bins=50)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

## Right-skewed

```python
df['class'].value_counts()
```

```python
df['class'].value_counts().plot(kind='pie',
                                autopct='%1.f%%')
plt.title("Price Distribution by Flight Class")
plt.show()
```

```python
df['airline'].value_counts()        
```

```python
df['airline'].value_counts().plot(kind='pie',
                                autopct='%1.f%%')
plt.title("Price Distribution by Airline")
plt.show()
```

```python
df['price'].hist()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='days_left', y='price', data=df, hue='class')
plt.title('Price Change')
plt.xlabel('Days remaining until departure')
plt.ylabel('Average Price')
plt.gca().invert_xaxis() 
plt.show()
```
